````python
# A função deve retornar a nota média (nota) das avaliações de locais de acampamento pertencentes ao país informado,
# considerando apenas as avaliações inseridas nos últimos p_qtde_dias dias anteriores à data atual (não inclui a data atual).

# Aaliacao - local - cidade - pais
sql_ex02 = """
DROP FUNCTION IF EXISTS media_avaliacoes;

CREATE FUNCTION media_avaliacoes(p_id_pais INT, p_qtde_dias INT)
RETURNS DECIMAL(4,2)
READS SQL DATA
BEGIN
    DECLARE media DECIMAL(4,2);
    DECLARE soma DECIMAL(4,2);
    DECLARE contagem INT;

    SELECT
        IFNULL(SUM(aval.nota), 0.00), IF(COUNT(aval.id_avaliacao)>0,COUNT(aval.id_avaliacao), 1) INTO soma, contagem
    FROM
        avaliacao AS aval
        JOIN local_acampamento AS loc ON (aval.id_local = loc.id_local)
        JOIN cidade AS cid ON (loc.id_cidade = cid.id_cidade)
        JOIN pais AS p ON (cid.id_pais = p.id_pais AND p.id_pais =p_id_pais)
    WHERE
        aval.data_avaliacao >= DATE_SUB(CURDATE(), INTERVAL p_qtde_dias DAY);

    SET media = soma/contagem;

    RETURN media;
END;
"""

# Faça seus testes no Workbench. Utilize o notebook apenas para submissão.
````

**a)** Crie uma função `categoria_frequencia` que recebe (nesta ordem):

- o RDD de notícias no formato acima;
- um inteiro `k`.

Considerando a frequência absoluta (contagem de ocorrências), a função deve retornar as `k` categorias mais frequentes no conjunto de notícias, seguida da sua respectiva frequência.

**Formato esperado de saída:**

- Uma lista de tuplas, onde cada tupla tem o formato `(categoria, frequencia)`.

**Obs**:

- Na saída, a categoria deve ser convertida para minúsculo.
- Ordene por frequência da maior para a menor.

````python
def categoria_frequencia(rdd_noticias, k):
    # "<ID_NOTICIA>,<CATEGORIA>,<AUTOR>,<DATA>,<FAKENEWS>,<TITULO>"
    total_noticias = rdd_noticias.count()
    separado = rdd_noticias.map(lambda x: x.split(','))
    categorias = separado.map(lambda x: (x[1].lower(),1))
    aparicoes = categorias.reduceByKey(lambda a,b: a+b)
    

    
    return aparicoes.takeOrdered(k, lambda x: -x[1])
    # k categorias mais frequentos no conjunto, respectiva frequenci

rdd_noticias_01 = sc.textFile("data/noticias/noticias_01.csv")

# Descomente caso queira executar localmente
print(categoria_frequencia(rdd_noticias_01, 10))
````

**b)** Crie uma função `resumo_categoria_noticias(rdd_noticias, categoria)` que recebe um RDD de notícias no formato descrito anteriormente e uma `categoria` em formato string.

Exemplo de chamada:

```python
resumo_categoria_noticias(rdd_noticias, "tecnologia")
```

A função deve retornar uma tupla no formato:

`(categoria, quantidade, quantidade_fakenews)`

Onde:
- `categoria`: nome da categoria da notícia;
- `quantidade`: total de notícias da categoria;
- `quantidade_fakenews`: considerando as notícias da categoria, quantas são fakenews.

**Exemplo de saída:**
`('tecnologia', 3, 1)`

**Obs e dicas**:
- A `categoria` pode ser fornecida qualquer padrão de maiúsculos e minúsculos;
- Em python, uma string que representa um inteiro pode ser convertida para inteiro utilizando a função `int()`. Por exemplo, `int("123")`.
- O retorno deve ser uma tupla, ou seja, o formato deve ser exatamente como o exemplo de saída apresentado acima, com os tipos corretos (string e inteiros).

````python
def resumo_categoria_noticias(rdd_noticias, categoria):
    # "<ID_NOTICIA>,<CATEGORIA>,<AUTOR>,<DATA>,<FAKENEWS>,<TITULO>"
    separado = rdd_noticias.map(lambda x: x.split(','))
    filtrada = separado.filter(lambda x: x[1].lower() == categoria.lower())
    total_cat = filtrada.count()
    total_fakes = filtrada.filter(lambda x: int(x[4])==1).count()
    return (categoria.lower(),total_cat,total_fakes)

rdd_noticias_01 = sc.textFile("data/noticias/noticias_01.csv")

# Descomente caso queira executar localmente
print(resumo_categoria_noticias(rdd_noticias_01, "jogos"))
````

<div class="alert alert-success">

Existem 3 fenômenos que podem ocorrer durante a leitura/observação de dados em databases, sendo estes: Dirty Read, Non-Repeatable Read e Phantom Read. Aplicar distintos níveis de isolamento garante quais destes eventos com certeza não vão ocorrer, mas essa garantia vem com um preço, geralmente atrelado a performance do serviço. A decisão de aplicar ou não um nível de isolamento exige compreensão do contexto, pois estes fenômenos podem representar um problema grave que deve ser evitado a todo custo ou irrelevante. 

O nível de isolamento Repeatable Read garante a impossibilidade de que ocorra Dirty Read e Non-Repeatable Read, mas não protege um serviço de Phantom Reads.

Aplicando-se Repeatable Read, caso uma transação faça COMMIT entre as leituras da transação em questão, é eliminada a chance de obter valores distintos da mesma linha, que poderia promover inconsistência dos dados observados.
