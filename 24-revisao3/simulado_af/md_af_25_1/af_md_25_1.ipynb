Essa tendência tem impactado negativamente a receita recorrente e aumentado os custos com aquisição de novos clientes. Preocupada com esse cenário, a empresa decidiu investir em uma solução baseada em dados para identificar com antecedência quais clientes estão propensos a abandonar a plataforma.

Para isso, o time de Data Science foi encarregado de desenvolver um modelo preditivo de churn, utilizando dados históricos de comportamento dos clientes. O objetivo é prever, com antecedência, a probabilidade de churn para cada cliente, permitindo ações proativas de retenção, como ofertas personalizadas, campanhas de engajamento e suporte direcionado.

No desenvolvimento do modelo, o time percebeu que precisará avaliar o desempenho do modelo em uma base de testes. Para isto, construiu um **CSV** contendo as colunas `id,churn,pred`, onde:

- `id`: é o identificador do cliente.
- `churn`: é o valor real (indica se o cliente deixou ou não de comprar com a empresa). `0` indica que o cliente continua a comprar (não teve **churn**), `1` indica um ex-cliente (teve **churn**).
- `pred`: é a probabilidade de churn predita pelo modelo desenvolvido pelo time.

Sua tarefa é criar uma função `avaliar_modelo(rdd, cutoff)` que recebe um **rdd** contendo os dados no formado do **CSV** e um valor de `cutoff`.

Em machine learning e estatística, `cutoff` (ou limiar) é o valor de corte usado para transformar uma probabilidade predita (como a coluna `pred` do CSV) em uma classe categórica (como  `churn = 0` ou `1`).

**Exemplo prático:**

Se você tem um modelo que prevê a probabilidade de um cliente abandonar um serviço (churn) e ele prevê 0.8 para um cliente, você precisa decidir a partir de que valor essa predição será considerada como “sim, ele vai sair”.

Se o `cutoff` for 0.5, então:

- `pred >= 0.5` → classifica como `churn = 1`

- `pred < 0.5` → classifica como `churn = 0`

**O que retornar**: considerando o `rdd` descrito e o `cutoff`, sua função deve calcular métricas de qualidade do modelo e retornar uma tupla no formato `(acuracia, precision, recall, f1_score)`:

- `acuracia`: faça da mesma forma como fez no projeto 2.

- `precision`: Considere a fórmula $$\text{Precision} = \frac{VP}{VP + FP}$$
    Sendo que:
    - `VP` – Verdadeiro Positivo: o modelo previu churn e o cliente realmente teve churn.
    - `VN` – Verdadeiro Negativo: o modelo previu não churn e o cliente realmente ficou.
    - `FP` – Falso Positivo: o modelo previu churn, mas o cliente não saiu.
    - `FN` – Falso Negativo: o modelo previu que o cliente ficaria, mas ele saiu.

- `recall`: $$\text{Recall} = \frac{VP}{VP + FN}$$

- `f1_score`: $$\text{F1-Score} = 2 \cdot \frac{\text{Precisão} \cdot \text{Recall}}{\text{Precisão} + \text{Recall}}$$



**Obs**:
