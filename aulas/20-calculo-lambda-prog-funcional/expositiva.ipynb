```python
from functools import reduce
```

```python
valores = [100, 200, 500, 30, 8700, 28000, 9, 75000]
```

```python
def calcular_desconto(valor):
    """Calcula valor descontado, se for maior que ou igual a 1000, aplica 10% de desconto"""
    if valor < 1000:
        return valor
    return valor * 0.9
```

```python
# testando
calcular_desconto(1000)
```

```python
# Fazer list comprehension para aplicar o desconto a cada valor
valor_com_desconto = [calcular_desconto(valor) for valor in valores]
valor_com_desconto
```

```python
# Agora utilizando a função map
valor_com_desconto2 = list(map(calcular_desconto, valores))
valor_com_desconto2
```

```python
def calcular_desconto(valor):
    """Calcula valor descontado, se for maior que ou igual a 1000, aplica 10% de desconto"""
    if valor < 1000:
        return valor
    return valor * 0.9

# Map com lambda function
valor_com_desconto2 = list(map(lambda valor: valor if valor < 1000 else valor * 0.9, valores))
valor_com_desconto2
```

```python
dobra = lambda x: x*2
dobra(10)
```

```python
# Filtre apenas os valores a partir de 1000 utilizando filter
valor_onde_aplico = list(filter(lambda x: x >= 1000, valores))
valor_onde_aplico
```

```python
# Qual o valor total dos itens utilizando reduce
valor_total = reduce(lambda x, y: x+y, valores)
valor_total
```

```python
# Exibir valores formatados
def calcular_reais(valores):
    def formata_reais(valor):
        return f"R$ {valor:.2f}"
    # Criar funcao que formata em reais
    return [formata_reais(v) for v in valores]

calcular_reais(valores)
```

```python

def formatar_reais(valor):
    return f"R$ {valor:.2f}"

def calcular_moeda(funcao_calculo, format_func):
    def calc(valores):
        return [format_func(funcao_calculo(v)) for v in valores]
    return calc

calcular_reais = calcular_moeda(calcular_desconto, formatar_reais)
# 
calcular_reais

calcular_reais(valores)
```
