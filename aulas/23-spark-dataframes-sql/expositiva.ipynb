```python
df = spark.read.csv("pacientes.csv", header=True, inferSchema=True)
df.show(5)
```

```python
from pyspark.sql.functions import col

df = df.withColumn("PatientId", col("PatientId").cast("long"))
```

```python
df.show(50)
```

```python
df.printSchema()
```

```python
spark.sql("DROP DATABASE IF EXISTS hospital")
```

```python
spark.sql("CREATE DATABASE IF NOT EXISTS hospital")
```

```python
df.write.mode("overwrite").saveAsTable("hospital.pacientes")
```

```python
%%time
df_pac = spark.read.table("hospital.pacientes")
df_pac.count()
```

```python
df_pac.show(3)
```

```python
df_pac.show(5)
```

```python
query = """
-- Contar pacientes por Gender
SELECT Gender,
        count(*) AS count
FROM hospital.pacientes
GROUP BY Gender
"""
df2 = spark.sql(query)
df2.show(5)
```

```python
from pyspark.sql.functions import col

df2 = (
    df_pac
    .groupBy(col("Gender"))
    .count()
)

df2.show()
```

```python
from pyspark.sql.functions import avg, col, max
import pyspark.sql.functions as F

df2 = (
    df_pac
    .groupBy(col("Gender"))
    .agg(
        F.avg(col("Age")).alias("AverageAge"),
        F.max(col("Age")).alias("MaisVelho")
    )
)

df2.show()
```
