# Sistemas de armazenamento

A escolha de sistemas de armazenamento é uma passo fundamental do gerenciamento eficiente em Big Data. Produzimos uma quantidade cada vez maior de dados diariamente, logo é imprescindível ter uma solução escalável e duradoura para armazenar tais informações, previnindo problemas tanto de performance quanto de proteção de dados sensíveis.

## RAID

RAID (**R**edundant **A**rray of **I**nexpensive **D**isks) é uma tecnologia de armazenamento que consiste na combinação de vários discos rígidos em um único sistema. O RAID pode ser configurado de várias maneiras diferentes, chamadas de níveis RAID, que oferecem diferentes benefícios e compromissos entre capacidade de armazenamento, velocidade de leitura/gravação e tolerância a falhas. O objetivo principal do RAID é fornecer um meio mais seguro e **confiável** de armazenar **grandes quantidades de dados**, com **redundância** para minimizar a perda de dados em caso de falha do hardware.

Vamos ver alguns níveis RAID.

### RAID `0`

O RAID 0 distribui os dados uniformemente em dois ou mais discos, sem informações de paridade, redundância ou tolerância a falhas. É normalmente usado para aumentar o desempenho, embora também possa ser usado como uma forma de criar um grande volume lógico a partir de dois ou mais discos físicos.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/9/9b/RAID_0.svg/330px-RAID_0.svg.png">

Fonte: https://upload.wikimedia.org/wikipedia/commons/thumb/9/9b/RAID_0.svg/330px-RAID_0.svg.png

Como o RAID 0 não fornece tolerância a falhas ou redundância, a falha de uma unidade fará com que todo o array falhe; como resultado da distribuição de dados em todos os discos, a falha resultará na perda total de dados.

### RAID `1`

O RAID 1 é um tipo de configuração de armazenamento em que os dados são copiados (produz *mirror* ou clone) para dois ou mais discos rígidos separados, oferecendo uma solução de backup imediata caso um dos discos falhe.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/b/b7/RAID_1.svg/330px-RAID_1.svg.png">

Fonte: https://upload.wikimedia.org/wikipedia/commons/thumb/b/b7/RAID_1.svg/330px-RAID_1.svg.png

Nesta configuração, as solicitações de leitura podem ser atendidas e tratadas por qualquer unidade na matriz. Dependendo da natureza da carga de E/S, o desempenho de leitura aleatória de uma matriz RAID 1 pode ser igual à soma do desempenho de cada membro, enquanto o desempenho de gravação permanece no nível de um único disco.

### RAID `5`

O RAID 5 é um tipo de configuração de armazenamento que utiliza **três ou mais** discos rígidos para oferecer uma solução de armazenamento com desempenho e tolerância a falhas. Nessa configuração, os dados são distribuídos por todos os discos rígidos, o que permite que a leitura e gravação de dados ocorra de forma mais rápida.

<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/f/fc/RAID5.png/330px-RAID5.png">

Fonte: https://upload.wikimedia.org/wikipedia/commons/thumb/f/fc/RAID5.png/330px-RAID5.png

Além disso, o RAID 5 possui um esquema de paridade distribuída, onde um bloco de paridade é armazenado em cada disco do conjunto. Isso significa que se um dos discos falhar, as informações contidas nele podem ser reconstruídas usando os dados armazenados nos outros discos e nas informações de paridade, garantindo a integridade dos dados e permitindo que o sistema continue operando sem perda de informações.



Veja mais em https://en.wikipedia.org/wiki/Standard_RAID_levels

**a)** Explique como você resolveria este problema. Cite tecnologias adequadas.

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

Para mover 10 PB de dados a 600 km de distância, a solução mais prática é o **transporte físico** dos HDs. O serviço equivalente no mercado é o **AWS Snowball Edge** (ou similares, como o Azure Data Box), que consiste em dispositivos robustos e lacrados para transporte seguro de grandes volumes de dados.

O processo seria:

2. Conectar os HDs ao novo servidor e copiar os dados para o storage definitivo.

Alternativas de rede seriam inviáveis: uma conexão de 10 Gbps levaria ~92 dias para transferir 10 PB.

**b)** Qual seria a latência da sua solução?

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

A latência é o tempo desde o início da operação até o primeiro dado chegar ao destino.

No transporte físico por caminhão (600 km a ~75 km/h):
- **Latência ≈ 8 horas** (tempo do trajeto)

Se fosse via rede de fibra óptica de alta velocidade, a latência seria muito menor (~alguns milissegundos para o primeiro byte chegar), mas o tempo total de transferência seria enorme.

**c)** Qual a largura de banda e a taxa de transferência da sua solução?

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

Considerando a solução de transporte físico (caminhão):

- **Total de dados**: 10.000 HDs × 1 TB = 10.000 TB = 10 PB = 10.240.000 GB
- **Tempo de trajeto**: 600 km ÷ 75 km/h = 8 horas = 28.800 segundos

**Taxa de transferência (throughput)**:
```
throughput = 10.240.000 GB / 28.800 s ≈ 355,6 GB/s
```

**Largura de banda**: não se aplica no sentido tradicional de rede, pois o meio de transmissão é o transporte físico. Porém, analogamente, podemos dizer que a "largura de banda efetiva" é de ~355,6 GB/s — muito superior a qualquer link de rede disponível comercialmente.

</div>

**d)** Suponha que você resolva colocar todos os HDs em um caminhão e disparar! Qual seria a latência e a taxa de transferência da solução?

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

Ignorando o tempo de carga/descarga dos HDs:

- **Velocidade média do caminhão**: ~75 km/h
- **Distância**: 600 km
- **Latência**: 600 / 75 = **8 horas**

**Taxa de transferência**:
```python
qtde_hds = 10000
gb = qtde_hds * 1 * 1024   # 10.000 TB → GB

horas = 8
segundos = horas * 60 * 60  # 28.800 s

throughput = gb / segundos  # ≈ 355,56 GB/s
```

**Resultado: latência ≈ 8 horas, taxa de transferência ≈ 355,56 GB/s**

Este é o famoso princípio *"Never underestimate the bandwidth of a station wagon full of tapes hurtling down the highway"* — Andrew Tanenbaum. O transporte físico frequentemente supera qualquer link de rede quando o volume de dados é muito grande.

**Exercício 2** (*Prova Final 2019-2*)

Em um cluster temos vários arquivos enormes, de tamanho médio 1 GB. Todos os arquivos são do tipo write-once-read-many: uma vez que o arquivo é criado, ele pode apenas ser removido por completo, nunca modificado parcialmente. Esse é o modo de operação do S3 da AWS. 
Os arquivos são divididos em blocos de 64 MB e estão armazenados de modo redundante, com fator de replicação 3 (este é o modo padrão do HDFS). Ou seja, cada bloco é armazenado 3 vezes, em máquinas diferentes. 
Desejamos armazenar 10000 arquivos destes, em máquinas com 4 discos de 1TB SSD cada, nas quais queremos que os dados ocupem no máximo 75% do espaço. 
Toda semana parte dos arquivos será processada usando Spark para alguma atividade de extração de informação. 

**a)** Quais as vantagens e desvantagens da replicação em máquinas diferentes, neste contexto? 

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

**Vantagens da replicação em máquinas diferentes:**
- **Tolerância a falhas (confiabilidade)**: se uma máquina falhar, os dados ainda existem em outras 2 máquinas. O sistema continua operando sem perda de dados.
- **Desempenho em leitura**: o Spark pode ler blocos em paralelo de réplicas diferentes, maximizando o throughput. A princípio, o scheduler do Spark tenta alocar a tarefa na máquina que já possui o dado localmente (*data locality*), reduzindo tráfego de rede.

**Desvantagens:**
- **Custo de armazenamento**: cada dado ocupa 3x mais espaço físico (fator de replicação 3).
- **Custo de escrita**: cada escrita precisa ser propagada para 3 máquinas diferentes, aumentando a latência de escrita e o tráfego de rede.
- **Custo de aquisição e operação**: mais máquinas = mais hardware, energia e manutenção.

</div>

**b)** Porque write-once-read-many é desejável aqui? 

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

O modelo *write-once-read-many* (WORM) é desejável porque:

1. **Simplicidade de replicação**: como o arquivo nunca é modificado parcialmente, as réplicas são sempre idênticas. Não há necessidade de sincronização de escrita entre réplicas — o que eliminaria a complexidade de protocolos de consistência distribuída (ex: Paxos, Raft).

2. **Coerência garantida**: qualquer réplica pode ser lida com confiança, pois todas são cópias exatas do arquivo original. Não existe o problema de leitura de dados "sujos" ou inconsistentes.

3. **Compatibilidade com pipelines de Big Data**: em Big Data é comum o padrão `dados brutos → pipeline Spark → dados processados`. Os dados brutos são escritos uma vez e lidos repetidamente por diferentes jobs sem necessidade de modificação.

4. **Simplicidade operacional**: remover um arquivo é atômico (deleta todas as réplicas), e nunca há "merge" de atualizações parciais.

</div>

**c)** Quantas máquinas este cluster deve ter, no mínimo, para acomodar esses dados?

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

**Cálculo:**

- 10.000 arquivos × 1 GB/arquivo = **10.000 GB** de dados originais
- Fator de replicação 3 → dados reais armazenados: 10.000 × 3 = **30.000 GB = 30 TB**
- Cada máquina tem: 4 discos × 1 TB = **4 TB por máquina**
- Limite de uso de 75%: cada máquina armazena no máximo 4 × 0,75 = **3 TB úteis por máquina**

**Número mínimo de máquinas:**
```
30.000 GB / 3.000 GB por máquina = 10 máquinas
```

**Resposta: mínimo de 10 máquinas.**

Obs.: o fator de replicação 3 também exige que cada bloco esteja em ao menos 3 máquinas distintas, portanto o mínimo absoluto para garantir isso também é 3 máquinas — mas o gargalo aqui é a capacidade, que exige 10.

</div>

**d)** Mais dados chegam constantemente ao sistema, que eventualmente ficará sem capacidade! Mas uma vez que a pipeline Spark é executada, os dados não são mais necessários de imediato. Proponha uma solução simples para não perder dados e não ter que aumentar indefinidamente o cluster. 

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

**Solução: política de ciclo de vida com armazenamento em camadas (tiered storage)**

Após a pipeline Spark processar os arquivos semanalmente, os dados processados já foram extraídos. Os arquivos brutos originais podem ser movidos para um storage mais barato e de menor disponibilidade imediata:

1. **Dados "quentes" (hot)**: arquivos recentes ainda não processados → ficam no HDFS do cluster (SSDs, alta velocidade).
2. **Dados "frios" (cold)**: arquivos já processados → são movidos para armazenamento de arquivamento barato, como **AWS S3 Glacier**, **fitas magnéticas (LTO)** ou **HDDs de baixo custo em cold storage**.

Implementação prática:
- Após cada execução do Spark, mover os arquivos processados para o Glacier via política de ciclo de vida do S3.
- O cluster HDFS mantém apenas os dados ainda não processados, liberando espaço continuamente.
- Se uma auditoria ou reprocessamento for necessário, os dados podem ser restaurados do Glacier (com latência de minutos a horas, o que é aceitável).

Esta abordagem evita crescimento indefinido do cluster sem perda de dados.

**Exercício 5** Considerando a capacidade total de todos os discos, qual a porcentagem aproximada realmente disponível para armazenamento no RAID 0, RAID 1 e RAID 5 (considere 4 discos)?

<div style="background-color: rgba(71, 85, 155, 0.2); padding: 10px; border-radius: 5px; border-left: 4px solid #475569; border: 1px solid rgba(71, 85, 105, 0.3);">

Considerando 4 discos de mesma capacidade **C** cada:

| RAID | Capacidade disponível | % do total |
|------|----------------------|------------|
| **RAID 0** | 4 × C = 4C | **100%** |
| **RAID 1** | 1 × C = C (todos os discos espelham o mesmo dado) | **25%** |
| **RAID 5** | (N-1) × C = 3C (1 disco reservado para paridade) | **75%** |

**Explicações:**

- **RAID 0**: striping puro, sem redundância. Todos os discos contribuem com 100% de sua capacidade para armazenamento. Não há espaço "desperdiçado" com paridade ou espelhamento. Porém, sem tolerância a falhas.

- **RAID 1**: espelhamento. Com 4 discos, metade (ou todos, em configurações aninhadas) carregam cópias exatas. Na versão clássica com 4 discos espelhados, disponibiliza apenas 1 disco de capacidade efetiva → **25%**.

- **RAID 5**: paridade distribuída. Com N discos, o equivalente a 1 disco é usado para paridade (distribuída entre todos). Com 4 discos: 3 discos de dados, 1 de paridade equivalente → **75%** disponível.
